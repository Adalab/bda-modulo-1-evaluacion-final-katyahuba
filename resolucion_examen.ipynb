```python
def agregar_producto(nombre, precio, cantidad):

    for producto in inventario:
        if producto["nombre"] == nombre:
            producto["cantidad"] += cantidad
            print("el producto ya existe, hemos actualizado la cantidad")
            return "se ha actualizado la cantidad"
        
    inventario.append({"nombre":nombre, "precio": precio, "cantidad": cantidad})
    print("el producto no lo teniamos, asi que lo agregamos")
```

```python
agregar_producto("short", 25, 5)
print(inventario)
```
